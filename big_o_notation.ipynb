```python
import numpy as np
import seaborn
import math
import matplotlib.pyplot as plt
```

```python
#Due to the functions being plotted, we want to see the plot for values of ~0-5
#Don't include 0 b/c log can't take 0
n = np.linspace(0.05,3,30,endpoint=True)
#Create separate x values for factorial, as it only handles integers
n_int = list(range(1,4))
#Y-values!
y1 = [(i*0 + 1) for i in n]
y2 = [math.log(i) for i in n]
y3 = n
y4 = [i*math.log(i) for i in n]
y5 = [i**2 for i in n]
y6 = [(2**i) for i in n]
y7 = [math.factorial(i) for i in n_int]
y8 = [i**i for i in n]
y = [y1,y2,y3,y4,y5,y6,y8]
lbl = ['O(1)','O(log n)','O(n)','O(n log n','O(n^2)','O(2^n)','O(n^n)']


plt.figure()
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
for i in range(len(y)):
    plt.plot(n,y[i],label=lbl[i])
plt.plot(n_int,y7,'x',label='O(n!)')    
plt.legend()
plt.ylim(0,10)
```
