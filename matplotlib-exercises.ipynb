```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#1) Plot y = x^2 + x +2

```python
#Create values
#list from -5 to 5, with 101 points and include 5 as last value
x = np.linspace(-5,5,101,endpoint=True)
y = x**2 - x + 2
y2 = [(n**2-n+2) for n in x]

plt.figure()
#fig, [ax1,ax2] = plt.subplots(1,2)
#plt.style.use('seaborn-darkgrid')
plt.style.use('default')
plt.subplot(121)
plt.plot(x,y)
plt.annotate('(0,0)',xy = (0,0),xytext = (-5,6), arrowprops={'width':2,'headwidth':8})
#plt.text(-1,1,'(0,0)')
plt.ylim(0,30)
plt.title('y = $x^2$ + x + 2')

plt.subplot(122)
plt.plot(x,y2)
plt.ylim(0,30)
plt.title('y = $x^2$ + x + 2\nusing list comprehension')

```

#2 & #3) graph: y=sqrt(x), y = x^3, y=2^x 

```python
x = np.linspace(0,10,101,endpoint = True)
y1 = x**.5
y2 = x**3
y3 = [(2**i) for i in x]
y4 = [(1/(i+1)) for i in x]
#lists of y, color and title
y = [y1,y2,y3,y4]
c = ['r','b','k','g']
ttl = ['$y = \sqrt{x}$','$y = x^3$','$y = 2^x$',r'$y = \frac{1}{x + 1}$']
lbl = ['$y = \sqrt{x}$','$y = x^3$','$y = 2^x$',r'$y = \frac{1}{x + 1}$'] #frac requires raw string
#create figure
plt.figure()
plt.style.use('default')
#create subplots in loop
for i in range(len(y)):
    plt.subplot(2,2,i+1)
    plt.plot(x,y[i],c[i],label = lbl[i])
    plt.grid()
    plt.title(ttl[i])
    plt.legend()

plt.tight_layout()
```
